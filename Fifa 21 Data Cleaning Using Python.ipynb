## Converting height and weight columns to numerical values

```python
# removing special charcters
df['Height'] = df['Height'].str.replace("'", ".").str.replace('"', "")
```

```python
df['Height']
```

```python
# changing the data type to float
df['Height']=df['Height'].astype('float')
```

```python
df['Height'].dtype
```

```python
# extracting the weight
df['Weight'] = df['Weight'].str[:-3]
df['Weight']
```

```python
# chaning data type to int
df['Weight']=df['Weight'].astype('int')
```

```python
# renaming the columns
df=df.rename(columns={'Height':'Height_in_meters','Weight':'Weight_in_lbs'})
```

## Extracting team and contract from 'Team & Contract'	Column

```python
df['Team & Contract']
```

```python
df['Team & Contract']=df['Team & Contract'].str.replace('\n','')
```

```python
df['Team & Contract']
```

```python
# Convert Team & Contract Column to str to clean it
df['Team & Contract'] = df['Team & Contract'].astype('str')

# Create Teams, Contract_Begins and Contract_Ends columns 
df['Teams'] = ""
df['Contract_Begins'] = ""
df['Contract_Ends'] = ""
df['Contract_Type'] = ""


for i in range(len(df['Team & Contract'])):
    if '~' in df.loc[i, 'Team & Contract']:
        df.loc[i, 'Teams']=df.loc[i, 'Team & Contract'][:-11]
        df.loc[i, 'Contract_Begins']=df.loc[i, 'Team & Contract'][-11:-7]
        df.loc[i, 'Contract_Ends']=df.loc[i, 'Team & Contract'][-4:]
        df.loc[i, 'Contract_Type']='Contract'
    elif 'Free' in df.loc[i, 'Team & Contract']:
        df.loc[i, 'Teams']=df.loc[i, 'Team & Contract'][:-4]
        df.loc[i, 'Contract_Begins']=np.nan
        df.loc[i, 'Contract_Ends']=np.nan
        df.loc[i, 'Contract_Type']='Free'
    else:
        c = df.loc[i, 'Team & Contract'].split(",")[0][-2:]
        if ((c == " 1") | (c == " 2") | (c == " 3") | (c == " 4") | (c == " 5") | (c == " 6") | (c ==" 7") | (c == " 8") | (c == " 9")):
            df.loc[i, 'Teams']= df.loc[i,'Team & Contract'][:-19]
            df.loc[i, 'Contract_Begins']= df.loc[i,'Team & Contract'][-12:-7]
            df.loc[i, 'Contract_Ends']=np.nan
            df.loc[i, 'Contract_Type']='On Loan'
        else:
            df.loc[i, 'Teams']= df.loc[i,'Team & Contract'][:-20]
            df.loc[i, 'Contract_Begins']= df.loc[i,'Team & Contract'][-12:-7]
            df.loc[i, 'Contract_Ends']=np.nan
            df.loc[i, 'Contract_Type']='On Loan'
```

# Converting 'Value', 'Wage' & 'Release Clause' into real numbers

```python
# Remove the currency symbol and 'M' & 'K' from (Value, Wage & Release Clause) columns
for column in df.loc[:,'Value':'Release Clause']:
    for i in range(len(df)):
        if ('M' in df.loc[i,column]) and ('.' in df.loc[i,column]):
            df.loc[i,column] = df.loc[i,column].replace('€', '').replace('.','').replace('M', '00000')
        elif ('M' in df.loc[i,column]) and ('.' not in df.loc[i,column]):
            df.loc[i,column] = df.loc[i,column].replace('€', '').replace('M', '000000')
        elif ('K' in df.loc[i,column]) and ('.' in df.loc[i,column]):
            df.loc[i,column] = df.loc[i,column].replace('€', '').replace('.','').replace('K', '00')
        elif ('K' in df.loc[i,column]) and ('.' not in df.loc[i,column]):
            df.loc[i,column] = df.loc[i,column].replace('€', '').replace('K', '000')
        else:
            df.loc[i,column] = df.loc[i,column].replace('€', '')
```
